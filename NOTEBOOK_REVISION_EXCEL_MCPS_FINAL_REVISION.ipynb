```python
from pathlib import Path
import pandas as pd
import re

CARPETA = Path("/content/drive/MyDrive/9. Gabs datos finales")
archivos = sorted(CARPETA.glob("*.xlsx"))

dfs = {} # Initialize the dictionary to store DataFrames

if not archivos:
    print(f"⚠️ No se encontraron archivos .xlsx en la carpeta: {CARPETA}")
    print("Por favor, verifica que la ruta sea correcta y que contenga archivos Excel.")
else:
    for archivo in archivos:
        nombre = archivo.stem
        nombre = re.sub(r'\W+', '_', nombre)  # reemplaza espacios y caracteres especiales
        dfs[nombre] = pd.read_excel(archivo) # Store DataFrame in the dfs dictionary

    print("DataFrames creados:")
    for nombre_df in dfs.keys():
        print("-", nombre_df)
```

```python
import pandas as pd
import numpy as np

#===========================================
# CONFIGURACIÓN
#===========================================

COLUMNAS = [
    "COD_MCP_RENIEC",
    "DEPARTAMENTO",
    "PROVINCIA",
    "DISTRITO",
    "MCP"
]

#===========================================
# LIMPIEZA GENERAL
#===========================================

def limpiar_texto(x):
    if pd.isna(x):
        return np.nan
    return (
        str(x)
        .strip()
        .upper()
    )

#===========================================
# CONSTRUIR HISTORIAL
#===========================================

lista = []
equivalencia_envios = []

for i, (nombre, df) in enumerate(dfs.items(), start=1):

    envio = f"{i:02d}"

    df = df.copy()

    # Normalizar textos
    for c in ["DEPARTAMENTO","PROVINCIA","DISTRITO","MCP"]:
        df[c] = df[c].apply(limpiar_texto)

    # Renombrar COUNT(1)
    if "COUNT(1)" in df.columns:
        df = df.rename(columns={"COUNT(1)":"CANTIDAD"})

    # Agregar identificadores
    df["ENVIO"] = envio
    df["ARCHIVO"] = nombre

    lista.append(df)

    equivalencia_envios.append({
        "ENVIO": envio,
        "ARCHIVO": nombre
    })

df_long = pd.concat(lista, ignore_index=True)

equivalencia_envios = pd.DataFrame(equivalencia_envios)

print(df_long.shape)
display(df_long.head())

display(equivalencia_envios)
```

```python
import pandas as pd
import numpy as np

#=========================================================
# 1. RENOMBRAR CANTIDAD
#=========================================================

df_long = df_long.copy()

if "COUNT(1)" in df_long.columns:
    df_long = df_long.rename(columns={"COUNT(1)": "CANTIDAD"})


#=========================================================
# 2. TABLA BASE
#=========================================================

base = (
    df_long[
        [
            "DEPARTAMENTO",
            "PROVINCIA",
            "MCP"
        ]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)


#=========================================================
# 3. AGREGAR COLUMNAS DE CADA ENVÍO
#=========================================================

df_consolidado = base.copy()

envios = sorted(df_long["ENVIO"].unique())

for envio in envios:

    aux = (
        df_long[df_long["ENVIO"] == envio]
        [
            [
                "DEPARTAMENTO",
                "PROVINCIA",
                "MCP",
                "COD_MCP_RENIEC",
                "DISTRITO",
                "CANTIDAD"
            ]
        ]
        .rename(columns={
            "COD_MCP_RENIEC": f"COD_{envio}",
            "DISTRITO": f"DIST_{envio}",
            "CANTIDAD": f"CANT_{envio}"
        })
    )

    df_consolidado = df_consolidado.merge(
        aux,
        how="left",
        on=["DEPARTAMENTO","PROVINCIA","MCP"]
    )


#=========================================================
# 4. COLUMNAS AUXILIARES
#=========================================================

cod_cols  = sorted([c for c in df_consolidado.columns if c.startswith("COD_")])
dist_cols = sorted([c for c in df_consolidado.columns if c.startswith("DIST_")])


# Último código conocido
df_consolidado["COD_ACTUAL"] = (
    df_consolidado[cod_cols]
    .ffill(axis=1)
    .iloc[:, -1]
)

# Último distrito conocido
df_consolidado["DIST_ACTUAL"] = (
    df_consolidado[dist_cols]
    .ffill(axis=1)
    .iloc[:, -1]
)

# Número de envíos
df_consolidado["N_ENVIOS"] = (
    df_consolidado[cod_cols]
    .notna()
    .sum(axis=1)
)

# Cambio de código
df_consolidado["CAMBIO_CODIGO"] = (
    df_consolidado[cod_cols]
    .nunique(axis=1, dropna=True)
)

# Cambio de distrito
df_consolidado["CAMBIO_DISTRITO"] = (
    df_consolidado[dist_cols]
    .nunique(axis=1, dropna=True)
)
```

```python
#=========================================================
# 5. ¿EL NOMBRE APARECE EN OTRAS PROVINCIAS?
#=========================================================

homonimos = (
    df_long
    .groupby(["DEPARTAMENTO","MCP"])
    .agg(
        N_PROVINCIAS=("PROVINCIA","nunique"),
        PROVINCIAS=("PROVINCIA",
                    lambda x: " | ".join(sorted(set(x))))
    )
    .reset_index()
)

df_consolidado = df_consolidado.merge(
    homonimos,
    on=["DEPARTAMENTO","MCP"],
    how="left"
)


#=========================================================
# 6. ¿ESTA MCP TUVO MÁS DE UN CÓDIGO?
#=========================================================

revision_codigos = (
    df_long
    .groupby(["DEPARTAMENTO","PROVINCIA","MCP"])
    .agg(
        N_CODIGOS=("COD_MCP_RENIEC","nunique"),
        CODIGOS=("COD_MCP_RENIEC",
                 lambda x: " | ".join(sorted(set(x.dropna()))))
    )
    .reset_index()
)

df_consolidado = df_consolidado.merge(
    revision_codigos,
    on=["DEPARTAMENTO","PROVINCIA","MCP"],
    how="left"
)
```

```python
def clasificar(row):

    if row["N_PROVINCIAS"] > 1:
        return "HOMONIMO"

    if row["N_CODIGOS"] > 1:
        return "CAMBIO_CODIGO"

    if row["CAMBIO_DISTRITO"] > 1:
        return "CAMBIO_DISTRITO"

    return "OK"


df_consolidado["OBSERVACION"] = (
    df_consolidado
    .apply(clasificar, axis=1)
)
```

```python
#=========================================================
# CLASIFICACIÓN DE OBSERVACIONES
#=========================================================

def observaciones(row):

    obs = []

    # El nombre aparece en más de una provincia
    if row["N_PROVINCIAS"] > 1:
        obs.append("HOMONIMO")

    # La misma MCP tuvo más de un código
    if row["N_CODIGOS"] > 1:
        obs.append("CAMBIO_CODIGO")

    # La misma MCP cambió de distrito
    if row["CAMBIO_DISTRITO"] > 1:
        obs.append("CAMBIO_DISTRITO")

    if len(obs) == 0:
        return "OK"

    return " | ".join(obs)


df_consolidado["OBSERVACIONES"] = (
    df_consolidado
    .apply(observaciones, axis=1)
)

df_consolidado["REQUIERE_REVISION"] = (
    df_consolidado["OBSERVACIONES"] != "OK"
)
```

```python
print("Resumen de incidencias\n")

print(df_consolidado["OBSERVACIONES"].value_counts())

print("\n")

print(
    f"MCP que requieren revisión: "
    f"{df_consolidado['REQUIERE_REVISION'].sum():,}"
)
```
